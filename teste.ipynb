```python
import pandas as pd

df = pd.read_csv("pizzas.csv")
df
```

# Explicando o gráfico

abaixo o gráfico para ver a relação do diametro da pizza com seu valor, podemos ver que, confome maior o diametro, maior o valor da pizza <br>
**corelação positiva e muito forte**

```python
df.plot(kind="line",color="#229a00", x="diametro", y="preco"); 
```

```python
from sklearn import linear_model

modelo = linear_model.LinearRegression()

x = df[["diametro"]]
y = df[["preco"]]

modelo.fit(x, y)
```

```python
print(modelo.predict([[12]])[0][0])
```
